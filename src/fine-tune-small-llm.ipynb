```python
def format_prompt(sample):
    prompt=f"""
        You are a virtual assistant for QA support. Just answer the proposed question by the user.\n\n
        Context:\n
        {sample['context']}\n
        Question:\n
        {sample['question']}\n
        Answer:\n
        {sample['answers']}      

    """
    return prompt
```
