```python
from datasets import load_dataset
```

Carregar um dataset de pergunta e resposta

```python
dataset = load_dataset('squad', split="train")
```

```python
train_dataset, eval_dataset = dataset.train_test_split(test_size=0.2).values()
print(f"Train dataset size: {len(train_dataset)}")
print(f"Eval dataset size: {len(eval_dataset)}")
```

```python
def format_prompt(sample):
    """Format SQuAD dataset for QA fine-tuning"""
    context = sample["context"]
    question = sample["question"]
    answer = sample["answers"]["text"][0]  # Get first answer
    
    prompt = f"""You are a helpful QA assistant. Answer the question based on the context provided.

Context: {context}

Question: {question}

Answer: {answer}"""
    
    return {"text": prompt}
```

Carregamento do modelo de base

```python
# Format datasets
train_dataset = train_dataset.map(format_prompt, remove_columns=train_dataset.column_names)
eval_dataset = eval_dataset.map(format_prompt, remove_columns=eval_dataset.column_names)
print("✓ Datasets formatted successfully!")
```

```python
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    TrainingArguments
)
```

```python
from unsloth import FastLanguageModel
from unsloth import unsloth_training_params

# Model configuration
max_seq_length = 2048
model_name = "unsloth/gemma-3-270m-it-unsloth-bnb-4bit"

# Load base model with 4-bit quantization
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=model_name,
    max_seq_length=max_seq_length,
    load_in_4bit=True,  # 4-bit quantization for efficiency
    load_in_8bit=False,
    dtype=None,  # Auto-detect
)
print(f"✓ Model loaded: {model_name}")
```

## Training Configuration and Execution

```python
from peft import LoraConfig, get_peft_model
from trl import SFTTrainer, SFTConfig

# Configure LoRA for efficient fine-tuning
model = FastLanguageModel.get_peft_model(
    model,
    r=16,  # LoRA rank
    lora_alpha=32,  # LoRA scaling
    lora_dropout=0.05,
    bias="none",
    use_gradient_checkpointing="unsloth",  # Use Unsloth's gradient checkpointing
    random_state=42,
)

print("✓ LoRA configuration applied")
```

```python
# Save the fine-tuned model
output_dir = "../results/fine-tuned-model"
model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)

# Also save 16-bit version for better compatibility
model.push_to_hub_gguf(
    repo_id="gemma-3-270m-qa",
    tokenizer=tokenizer,
    quantization_method="q4_k_m",
    folder_name=f"{output_dir}/gguf",
)

print(f"✓ Model saved to {output_dir}")
```
